```python
# Write a program to implement Hangman game using python.
# Description:
# Hangman is a classic word-guessing game. The user should guess the word correctly by
# entering alphabets of the user choice. The Program will get input as single alphabet from the
# user and it will matchmaking with the alphabets in the original

import random

name = input("What is your name? ")
print(f"Good Luck, {name}!")

words = ['rainbow', 'computer', 'science', 'programming',
         'python', 'mathematics', 'player', 'condition',
         'reverse', 'water', 'board', 'geeks']

word = random.choice(words)
print("\nGuess the word!")

guesses = ""
turns = 12

while turns > 0:
    display = [char if char in guesses else "_" for char in word]
    print(" ".join(display))

    if "_" not in display:
        print("\nYou Win!")
        print(f"The word is: {word}")
        break

    guess = input("\nGuess a character: ").lower()

    if guess in guesses:
        print("You already guessed that letter!")
        continue

    guesses += guess

    if guess not in word:
        turns -= 1
        print("Wrong!")
        print(f"You have {turns} guesses left.")

        if turns == 0:
            print("\nYou Lose!")
            print(f"The word was: {word}")
```

```python
# Write a Python program to implement A* algorithm. Refer the following graph as an Input for
# the program.

from queue import PriorityQueue


graph = {
    'A': [('B', 2), ('E', 3)],
    'B': [('A', 2), ('F', 9),('C', 1)],
    'C': [('B', 1)],
    'D': [('E', 6),('F', 1)],
    'E': [('A', 3), ('D', 6)],
    'F': [('B', 9), ('D', 1)],
    'G': []  
}


heuristic = {
    'A': 11,
    'B': 6,
    'C': 99,
    'D': 1,
    'E': 7,
    'F': 0,
}

def a_star(start, goal):
    open_set = PriorityQueue()
    open_set.put((0, start))
    g_cost = {node: float('inf') for node in graph}
    g_cost[start] = 0
    came_from = {start: None}

    while not open_set.empty():
        current_f_score, current = open_set.get()

        if current == goal:
            path = []
            while current:
                path.append(current)
                current = came_from[current]
            return path[::-1]

        for neighbor, cost in graph[current]:
            tentative_g_score = g_cost[current] + cost

            if tentative_g_score < g_cost[neighbor]:
                came_from[neighbor] = current
                g_cost[neighbor] = tentative_g_score
                f_score = tentative_g_score + heuristic[neighbor]
                open_set.put((f_score, neighbor))

    return None


start_node = 'A'
goal_node = 'F'
path = a_star(start_node, goal_node)
print("Path from {} to {}: {}".format(start_node, goal_node, path))
```
